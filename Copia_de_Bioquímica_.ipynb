```python
import sqlite3

# Crear o conectar la base de datos
conn = sqlite3.connect('compuestos_salud.db')
cursor = conn.cursor()

# Crear la tabla
cursor.execute('''CREATE TABLE IF NOT EXISTS compuestos (
                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                    nombre TEXT NOT NULL,
                    alimentos TEXT NOT NULL,
                    funcion TEXT NOT NULL,
                    ingesta_recomendada TEXT
                )''')

# Insertar los datos en la tabla
compuestos = [
    ("Quercetina", "Cebolla, Manzanas, Uvas", "Antioxidante, Antiinflamatorio", "10-50 mg/día"),
    ("Kaempferol", "Col rizada, Espinaca, Brócoli", "Antioxidante, Anticancerígeno", "10-20 mg/día"),
    ("Naringenina", "Pomelo, Naranjas", "Antioxidante, Mejora la resistencia celular", "50 mg/día"),
    ("Epicatequina", "Té verde, Cacao, Manzanas", "Antioxidante, Mejora la salud cardiovascular", "200-400 mg/día"),
    ("Epigalocatequina", "Té verde", "Antioxidante, Anticancerígeno", "300-400 mg/día"),
    ("Hesperidina", "Naranjas, Limones", "Antioxidante, Mejora la circulación", "50-100 mg/día"),
    ("Apigenina", "Perejil, Apio, Manzanilla", "Antiinflamatorio, Anticancerígeno", "10-20 mg/día"),
    ("Luteolina", "Pimientos verdes, Tomates, Apio", "Antioxidante, Neuroprotector", "10-50 mg/día"),
    ("Genisteína", "Soja, Tofu, Edamame", "Antioxidante, Regulador hormonal", "30-50 mg/día"),
    ("Daidzeína", "Soja, Tofu, Edamame", "Antioxidante, Regulador hormonal", "30-50 mg/día"),
    ("Resveratrol", "Uvas rojas, Vino tinto", "Antioxidante, Antienvejecimiento", "5-30 mg/día"),
    ("Catequina", "Té verde, Cacao, Vino tinto", "Antioxidante, Mejora la salud metabólica", "200-400 mg/día"),
    ("Galato de epicatequina", "Té verde", "Antioxidante, Antiinflamatorio", "300-400 mg/día"),
    ("Galato de epigalocatequina", "Té verde", "Antioxidante, Neuroprotector", "300-400 mg/día"),
    ("Ácido gálico", "Moras, Té negro, Vino tinto", "Antioxidante, Antimicrobiano", "50-200 mg/día"),
    ("Ácido elágico", "Frambuesas, Fresas, Granadas", "Antioxidante, Anticancerígeno", "30-50 mg/día"),
    ("Rutina", "Alforfón, Cítricos", "Antioxidante, Mejora la circulación", "10-50 mg/día"),
    ("Silimarina", "Cardo mariano", "Hepatoprotector, Antioxidante", "200-400 mg/día"),
    ("Ácido cafeico", "Café, Hierbas, Cereales", "Antioxidante, Antiinflamatorio", "200-500 mg/día"),
    ("Ácido clorogénico", "Café, Alcachofas, Peras", "Antioxidante, Antihipertensivo", "200-400 mg/día"),
    ("Ácido ferúlico", "Salvado de trigo, Arroz integral", "Antioxidante, Antiinflamatorio", "200-500 mg/día"),
    ("Ácido p-cumárico", "Cacahuetes, Tomates, Cereales", "Antioxidante, Anticancerígeno", "50-200 mg/día"),
    ("Ácido sinápico", "Mostaza, Salvado de arroz", "Antioxidante, Antiinflamatorio", "50-200 mg/día"),
    ("Curcumina", "Cúrcuma", "Antiinflamatorio, Anticancerígeno", "500-1500 mg/día"),
    ("Tirosol", "Aceitunas, Aceite de oliva", "Antioxidante, Protector cardiovascular", "10-20 mg/día"),
    ("Hidroxitirosol", "Aceitunas, Aceite de oliva", "Antioxidante, Antiinflamatorio", "10-50 mg/día"),
    ("Oleuropeína", "Aceitunas, Aceite de oliva", "Antioxidante, Mejora la salud cardiovascular", "10-30 mg/día"),
    ("Procianidinas", "Cacao, Uvas, Manzanas", "Antioxidante, Mejora la salud vascular", "100-500 mg/día"),
    ("Antocianinas", "Arándanos, Moras, Cerezas", "Antioxidante, Mejora la visión", "50-200 mg/día"),
    ("Pelargonidina", "Fresas, Rábanos", "Antioxidante, Mejora la salud cardiovascular", "50-100 mg/día"),
    ("Cianidina", "Cerezas, Moras, Frambuesas", "Antioxidante, Antiinflamatorio", "50-100 mg/día"),
    ("Delphinidina", "Arándanos, Grosellas negras", "Antioxidante, Neuroprotector", "50-100 mg/día"),
    ("Peonidina", "Arándanos, Uvas rojas", "Antioxidante, Mejora la función inmune", "50-100 mg/día"),
    ("Malvidina", "Uvas rojas, Vino tinto", "Antioxidante, Protector cardiovascular", "50-100 mg/día"),
    ("Fisetina", "Fresas, Manzanas, Caquis", "Antioxidante, Neuroprotector", "20-50 mg/día"),
    ("Isorhamnetina", "Peras, Almendras, Pistachos", "Antioxidante, Antiinflamatorio", "10-50 mg/día"),
    ("Baicalina", "Raíz de escutelaria", "Antiinflamatorio, Neuroprotector", "200-400 mg/día"),
    ("Crisina", "Miel, Propóleo", "Antioxidante, Mejora el sueño", "500-1000 mg/día"),
    ("Galangina", "Jengibre, Galanga", "Antioxidante, Antimicrobiano", "50-200 mg/día"),
    ("Licopeno", "Tomates, Sandía, Papaya", "Antioxidante, Protector de la piel", "5-15 mg/día"),
    ("Astaxantina", "Algas, Camarones, Salmón", "Antioxidante, Mejora la salud ocular", "2-12 mg/día"),
    ("Capsaicina", "Chiles, Pimientos", "Antiinflamatorio, Mejora el metabolismo", "10-30 mg/día"),
    ("Zingibereno", "Jengibre", "Antiinflamatorio, Mejora la digestión", "50-100 mg/día"),
    ("Piperina", "Pimienta negra", "Antioxidante, Mejora la biodisponibilidad", "5-15 mg/día"),
    ("Thymol", "Tomillo, Orégano", "Antioxidante, Antimicrobiano", "10-50 mg/día"),
    ("Carvacrol", "Orégano, Mejorana", "Antioxidante, Antimicrobiano", "10-50 mg/día"),
    ("Eugenol", "Clavos de olor, Canela", "Antioxidante, Antiinflamatorio", "5-10 mg/día"),
    ("Rosmarinol", "Romero", "Antioxidante, Neuroprotector", "100-200 mg/día"),
    ("Ácido rosmarínico", "Romero, Menta, Albahaca", "Antioxidante, Antiinflamatorio", "100-200 mg/día"),
    ("Ácido ursólico", "Manzanas, Romero", "Antiinflamatorio, Anticancerígeno", "50-100 mg/día"),
    ("Ácido oleanólico", "Aceitunas, Romero", "Antioxidante, Protector hepático", "50-100 mg/día"),
    ("Apigenina-7-O-glucósido", "Manzanilla", "Antioxidante, Mejora el sueño", "10-20 mg/día"),
    ("Baicalina-7-O-glucósido", "Raíz de escutelaria", "Neuroprotector, Antiinflamatorio", "200-400 mg/día"),
    ("Diosmetina", "Cítricos", "Antioxidante, Mejora la circulación", "10-20 mg/día"),
    ("Tangeretina", "Cáscaras de naranja", "Antioxidante, Anticancerígeno", "50-100 mg/día"),
    ("Nobiletina", "Cáscaras de cítricos", "Antioxidante, Protector cardiovascular", "50-100 mg/día"),
    ("Morina", "Hojas de guanábana", "Antioxidante, Antimicrobiano", "50-100 mg/día"),
    ("Chalcona", "Lúpulo, Trébol", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Xantohumol", "Lúpulo", "Antioxidante, Mejora el metabolismo", "10-20 mg/día"),
    ("Aurapteno", "Cítricos", "Antioxidante, Anticancerígeno", "50-100 mg/día"),
    ("Umbeliferona", "Zanahorias, Apio", "Antioxidante, Antiinflamatorio", "10-50 mg/día"),
    ("Scopoletina", "Raíz de cúrcuma, Noni", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Psoraleno", "Higos, Perejil", "Antioxidante, Mejora la piel", "10-20 mg/día"),
    ("Bergapteno", "Pomelo, Cítricos", "Antioxidante, Antimicrobiano", "10-50 mg/día"),
    ("Resorcinol", "Almendras, Trigo", "Antioxidante, Mejora la piel", "5-15 mg/día"),
    ("Fisetina-3-O-glucósido", "Fresas, Uvas", "Neuroprotector, Antioxidante", "10-30 mg/día"),
    ("Isoxantohumol", "Lúpulo", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Artepillin C", "Propóleo", "Antimicrobiano, Mejora la inmunidad", "200-500 mg/día"),
    ("Pinocembrina", "Miel, Propóleo", "Antioxidante, Neuroprotector", "10-30 mg/día"),
    ("Ácido salvianólico", "Salvia, Romero", "Antioxidante, Mejora la salud cardiovascular", "200-400 mg/día"),
    ("Catequina-3-O-galato", "Té verde", "Antioxidante, Antiinflamatorio", "200-400 mg/día"),
    ("Gallocatequina", "Té verde", "Antioxidante, Neuroprotector", "200-400 mg/día"),
    ("Prodelphinidinas", "Uvas, Arándanos", "Antioxidante, Mejora la salud vascular", "100-300 mg/día"),
    ("Gallotannino", "Mango, Té verde", "Antioxidante, Antimicrobiano", "100-200 mg/día"),
    ("Elagitannino", "Granadas, Fresas", "Antioxidante, Anticancerígeno", "100-300 mg/día"),
    ("Punicalagina", "Granadas", "Antioxidante, Mejora la salud cardiovascular", "50-200 mg/día"),
    ("Ácido cinámico", "Canela, Frutas", "Antioxidante, Antimicrobiano", "20-50 mg/día"),
    ("Ácido coumárico", "Cereales, Tomates", "Antioxidante, Antiinflamatorio", "50-100 mg/día"),
    ("Eriodictiol", "Cítricos", "Antioxidante, Mejora la salud respiratoria", "20-50 mg/día"),
    ("Licochalcona A", "Raíz de regaliz", "Antioxidante, Antibacteriano", "50-100 mg/día"),
    ("Hesperidina", "Cáscaras de naranja, Limón", "Antioxidante, Mejora la circulación", "500-1000 mg/día"),
    ("Daidzeína", "Soja, Tofu", "Antioxidante, Fitoestrógeno", "20-50 mg/día"),
    ("Genisteína", "Soja, Edamame", "Antioxidante, Fitoestrógeno", "20-50 mg/día"),
    ("Formononetina", "Trébol rojo, Legumbres", "Fitoestrógeno, Mejora la salud ósea", "10-30 mg/día"),
    ("Biochanina A", "Trébol rojo, Soja", "Antioxidante, Fitoestrógeno", "10-30 mg/día"),
    ("Secoisolariciresinol", "Lino, Sésamo", "Antioxidante, Protector hormonal", "50-100 mg/día"),
    ("Matairesinol", "Lino, Sésamo", "Antioxidante, Anticancerígeno", "10-30 mg/día"),
    ("Enterodiol", "Lino, Frutas secas", "Antioxidante, Fitoestrógeno", "10-30 mg/día"),
    ("Enterolactona", "Lino, Cereales integrales", "Antioxidante, Mejora la salud hormonal", "10-30 mg/día"),
    ("Punicacorteína", "Corteza de granada", "Antioxidante, Antimicrobiano", "50-100 mg/día"),
    ("Tilirosida", "Tilo, Frutas", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Apigenina-6,8-C-diglucósido", "Cilantro, Perejil", "Antioxidante, Mejora la digestión", "10-30 mg/día"),
    ("Espectabilona", "Orquídeas", "Antioxidante, Antimicrobiano", "10-50 mg/día"),
    ("Baicalina-6-O-glucósido", "Escutelaria", "Antioxidante, Neuroprotector", "50-200 mg/día"),
    ("Ácido gentísico", "Cítricos, Fresas", "Antioxidante, Protector cardiovascular", "20-50 mg/día"),
    ("Ácido vaníllico", "Vainilla, Café", "Antioxidante, Mejora la salud hepática", "20-50 mg/día"),
    ("Furanoespiroetanoles", "Almendras, Ciruelas", "Antioxidante, Mejora la piel", "5-20 mg/día"),
    ("Amentoflavona", "Ginkgo biloba, Plantas medicinales", "Antioxidante, Mejora la memoria", "50-100 mg/día"),
    ("Crisofilán", "Jazmín", "Antioxidante, Mejora la salud respiratoria", "10-30 mg/día"),
    ("Carnosol", "Romero, Salvia", "Antioxidante, Mejora la salud cardiovascular", "50-200 mg/día"),
    ("Ácido caféico", "Café, Frutas", "Antioxidante, Antiinflamatorio", "200-400 mg/día"),
    ("Ácido ferúlico", "Arroz, Salvado", "Antioxidante, Mejora la salud de la piel", "50-200 mg/día"),
    ("Curcumina", "Cúrcuma", "Antioxidante, Antiinflamatorio", "500-1000 mg/día"),
    ("Resveratrol", "Uvas, Vino tinto", "Antioxidante, Protector cardiovascular", "5-20 mg/día"),
    ("Quercetina-3-O-glucósido", "Manzanas, Cebolla", "Antioxidante, Antiinflamatorio", "10-50 mg/día"),
    ("Kaempferol-3-O-rutinósido", "Brócoli, Espinacas", "Antioxidante, Protector hepático", "10-50 mg/día"),
    ("Rutinósido", "Cáscaras de cítricos", "Antioxidante, Mejora la circulación", "50-100 mg/día"),
    ("Isoflavonas", "Soja, Trébol rojo", "Fitoestrógeno, Mejora la salud ósea", "20-50 mg/día"),
    ("Fitoalexinas", "Plantaciones de uvas", "Antioxidante, Antimicrobiano", "10-30 mg/día"),
    ("Piceatannol", "Uvas, Maní", "Antioxidante, Antiinflamatorio", "5-20 mg/día"),
    ("Geraniol", "Aceite esencial de rosa", "Antioxidante, Mejora el estado de ánimo", "20-50 mg/día"),
    ("Ácido gálico", "Té, Uvas", "Antioxidante, Anticancerígeno", "100-300 mg/día"),
    ("Epigallo catequina-3-galato", "Té verde", "Antioxidante, Protector cardiovascular", "200-400 mg/día"),
    ("Proantocianidinas", "Uvas, Arándanos", "Antioxidante, Mejora la salud renal", "100-300 mg/día"),
    ("Hidroxitirosol", "Aceite de oliva", "Antioxidante, Protector cardiovascular", "5-10 mg/día"),
    ("Oleuropeína", "Aceitunas", "Antioxidante, Antimicrobiano", "50-100 mg/día"),
    ("Apigenina-7-O-rutinósido", "Manzanilla", "Antioxidante, Mejora el sueño", "10-20 mg/día"),
    ("Silibinina", "Cardo mariano", "Antioxidante, Protector hepático", "50-200 mg/día"),
    ("Xantona", "Mangostán", "Antioxidante, Anticancerígeno", "10-30 mg/día"),
    ("Mangiferina", "Hojas de mango", "Antioxidante, Mejora la salud ocular", "50-100 mg/día"),
    ("Luteolina", "Pimientos, Perejil", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Ácido rosmarínico", "Romero, Albahaca", "Antioxidante, Antimicrobiano", "50-200 mg/día"),
    ("Cianidina-3-O-glucósido", "Moras, Cerezas", "Antioxidante, Mejora la salud vascular", "50-100 mg/día"),
    ("Delphinidina-3-O-rutinósido", "Arándanos, Grosellas", "Antioxidante, Protector ocular", "50-150 mg/día"),
    ("Pelargonidina", "Fresas, Frambuesas", "Antioxidante, Mejora la piel", "30-100 mg/día"),
    ("Petunidina", "Uvas negras, Grosellas", "Antioxidante, Antiinflamatorio", "30-100 mg/día"),
    ("Malvidina", "Vino tinto, Uvas", "Antioxidante, Protector cardiovascular", "20-50 mg/día"),
    ("Peonidina", "Flores, Frutas rojas", "Antioxidante, Mejora la salud renal", "10-30 mg/día"),
    ("Siringina", "Sauces, Granos enteros", "Antioxidante, Mejora la piel", "20-50 mg/día"),
    ("Catequina-3-galato", "Té verde", "Antioxidante, Mejora la salud cerebral", "100-200 mg/día"),
    ("Teaflavinas", "Té negro", "Antioxidante, Mejora el metabolismo", "50-150 mg/día"),
    ("Flavonolignanos", "Cardo mariano", "Antioxidante, Protector hepático", "50-200 mg/día"),
    ("Lignanos", "Semillas de lino, Sésamo", "Antioxidante, Fitoestrógeno", "50-100 mg/día"),
    ("Aurona", "Árbol de aurones", "Antioxidante, Protector ocular", "10-30 mg/día"),
    ("Polidatina", "Uvas, Maní", "Antioxidante, Antiinflamatorio", "5-20 mg/día"),
    ("Tannatos", "Castañas, Té", "Antioxidante, Mejora la digestión", "50-100 mg/día"),
    ("Eugenol", "Clavo, Canela", "Antioxidante, Antimicrobiano", "5-15 mg/día"),
    ("Galangina", "Jengibre, Alpinia", "Antioxidante, Protector hepático", "20-50 mg/día"),
    ("Hidroxibenzaldehído", "Canela", "Antioxidante, Antimicrobiano", "5-20 mg/día"),
    ("Chalconas", "Manzanas, Tomates", "Antioxidante, Antiinflamatorio", "10-30 mg/día"),
    ("Kaempferol-7-O-glucósido", "Almendras, Espinacas", "Antioxidante, Mejora la circulación", "10-30 mg/día"),
    ("Fisetina", "Fresas, Manzanas", "Antioxidante, Protector neurológico", "10-50 mg/día"),
    ("Sakuraferina", "Flores de cerezo", "Antioxidante, Mejora la piel", "5-20 mg/día"),
    ("Miricetina", "Nueces, Té", "Antioxidante, Protector cardiovascular", "50-150 mg/día"),
    ("Tilirosida-7-O-glucósido", "Tilo, Moras", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Ácido elágico", "Frutas del bosque, Nueces", "Antioxidante, Anticancerígeno", "30-100 mg/día"),
    ("Geranilgeraniol", "Aceite de rosa, Cítricos", "Antioxidante, Antimicrobiano", "10-30 mg/día"),
    ("Aurapteno", "Cítricos, Pomelos", "Antioxidante, Protector cardiovascular", "20-50 mg/día"),
    ("Licopeno", "Tomates, Sandías", "Antioxidante, Protector de la próstata", "6-15 mg/día"),
    ("Betacaroteno", "Zanahorias, Calabaza", "Antioxidante, Protector ocular", "4-10 mg/día"),
    ("Astaxantina", "Algas, Salmón", "Antioxidante, Protector de la piel", "4-12 mg/día"),
    ("Zeaxantina", "Maíz, Espinacas", "Antioxidante, Protector ocular", "2-10 mg/día"),
    ("Luteína", "Espinacas, Brócoli", "Antioxidante, Protector ocular", "6-12 mg/día"),
    ("Bromelina", "Piña", "Antioxidante, Mejora la digestión", "500-1000 mg/día"),
    ("Tirosol", "Aceitunas, Aceite de oliva", "Antioxidante, Protector cardiovascular", "10-30 mg/día"),
    ("Naringina", "Pomelo, Naranjas", "Antioxidante, Mejora la circulación", "20-50 mg/día"),
    ("Galato de epicatequina", "Té verde", "Antioxidante, Mejora el metabolismo", "50-200 mg/día"),
    ("Piralgina", "Té negro, Canela", "Antioxidante, Antiinflamatorio", "10-30 mg/día"),
    ("Eriodictiol", "Cítricos, Hierbas", "Antioxidante, Mejora la salud respiratoria", "10-30 mg/día"),
    ("Crisina", "Miel, Propóleos", "Antioxidante, Fitoestrógeno", "10-50 mg/día"),
    ("Genisteína", "Soja, Frijoles", "Fitoestrógeno, Mejora la salud ósea", "20-50 mg/día"),
    ("Daidzeína", "Soja, Tofu", "Fitoestrógeno, Antioxidante", "20-50 mg/día"),
    ("Biochanina A", "Trébol rojo, Legumbres", "Fitoestrógeno, Antiinflamatorio", "10-30 mg/día"),
    ("Formononetina", "Trébol rojo, Legumbres", "Fitoestrógeno, Protector cardiovascular", "10-30 mg/día"),
    ("Silibina", "Cardo mariano", "Antioxidante, Protector hepático", "50-200 mg/día"),
    ("Ácido chicórico", "Achicoria, Albahaca", "Antioxidante, Mejora el sistema inmune", "50-150 mg/día"),
    ("Ácido quinico", "Café, Frutas", "Antioxidante, Mejora la digestión", "50-150 mg/día"),
    ("Hesperidina", "Cítricos", "Antioxidante, Mejora la circulación", "50-200 mg/día"),
    ("Diosmina", "Cáscaras de cítricos", "Antioxidante, Mejora la salud vascular", "50-150 mg/día"),
    ("Rhoifolina", "Cítricos", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Eupatorina", "Hierbas medicinales", "Antioxidante, Anticancerígeno", "10-30 mg/día"),
    ("Baicalina", "Scutellaria baicalensis", "Antioxidante, Protector neurológico", "20-50 mg/día"),
    ("Licochalcona A", "Regaliz", "Antioxidante, Antiinflamatorio", "10-30 mg/día"),
    ("Pinocembrina", "Propóleo, Miel", "Antioxidante, Mejora la salud cardiovascular", "10-30 mg/día"),
    ("Ácido clorogénico", "Café verde, Papas", "Antioxidante, Mejora el metabolismo", "50-200 mg/día"),
    ("Saligenina", "Sauces", "Antioxidante, Alivio del dolor", "10-30 mg/día"),
    ("Ácido protocatecuico", "Arándanos, Uvas", "Antioxidante, Protector hepático", "10-50 mg/día"),
    ("Equol", "Productos fermentados de soja", "Fitoestrógeno, Mejora la salud hormonal", "10-20 mg/día"),
    ("Ácido shikímico", "Anís estrellado", "Antioxidante, Mejora la inmunidad", "10-30 mg/día"),
    ("Isoxantohumol", "Lúpulo, Cerveza", "Fitoestrógeno, Mejora el sueño", "10-20 mg/día"),
    ("Amentoflavona", "Ginkgo biloba, Cedro", "Antioxidante, Protector cerebral", "20-50 mg/día"),
    ("Hinokiflavona", "Ginkgo, Coníferas", "Antioxidante, Antiinflamatorio", "10-30 mg/día"),
    ("Lignanos de pino", "Pino, Corteza", "Antioxidante, Mejora la piel", "30-50 mg/día"),
    ("Nobiletina", "Mandarinas, Cítricos", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Tangeretina", "Cítricos", "Antioxidante, Mejora la salud cardiovascular", "20-50 mg/día"),
    ("Ácido ferúlico", "Arroz, Trigo", "Antioxidante, Protector dérmico", "20-50 mg/día"),
    ("Gallocatequina", "Té verde, Frutas", "Antioxidante, Mejora el metabolismo", "50-200 mg/día"),
    ("Resorcinol", "Almendras, Col", "Antioxidante, Mejora la piel", "5-15 mg/día"),
    ("Proantocianidinas", "Uvas, Arándanos", "Antioxidante, Mejora la circulación", "50-150 mg/día"),
    ("Epicatequina galato", "Té verde", "Antioxidante, Protector cardiovascular", "100-200 mg/día"),
    ("Teagallina", "Té negro", "Antioxidante, Mejora la salud celular", "50-100 mg/día"),
    ("Hidroxitirosol", "Aceitunas, Aceite de oliva", "Antioxidante, Protector cardiovascular", "5-15 mg/día"),
    ("Oleuropeína", "Aceitunas", "Antioxidante, Mejora la salud inmune", "20-50 mg/día"),
    ("Secoisolariciresinol", "Lino, Sésamo", "Antioxidante, Fitoestrógeno", "20-50 mg/día"),
    ("Siringina", "Sauces, Granos", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Ácido p-cumárico", "Manzanas, Cerezas", "Antioxidante, Antimicrobiano", "20-50 mg/día"),
    ("Esculetina", "Castaño de Indias", "Antioxidante, Mejora la salud vascular", "10-30 mg/día"),
    ("Ácido cafeico", "Café, Frutas", "Antioxidante, Mejora el metabolismo", "50-150 mg/día"),
    ("Apigenina-7-glucósido", "Manzanilla, Perejil", "Antioxidante, Protector celular", "10-30 mg/día"),
    ("Cinarina", "Alcachofa", "Antioxidante, Protector hepático", "20-50 mg/día"),
    ("Ácido vanílico", "Cereales, Moras", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Ácido gentísico", "Morera, Granos", "Antioxidante, Mejora la circulación", "10-30 mg/día"),
    ("Ácido gálico", "Té, Uvas", "Antioxidante, Antiinflamatorio", "50-100 mg/día"),
    ("Metil gallato", "Té, Uvas", "Antioxidante, Antibacteriano", "10-30 mg/día"),
    ("Punicagalina", "Granada", "Antioxidante, Mejora la salud cardiovascular", "50-100 mg/día"),
    ("Ellagitanninos", "Frutas del bosque, Granada", "Antioxidante, Antiinflamatorio", "50-150 mg/día"),
    ("Malvidina", "Uvas rojas, Arándanos", "Antioxidante, Mejora la circulación", "10-30 mg/día"),
    ("Peonidina", "Uvas, Cerezas", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Pelargonidina", "Fresas, Frambuesas", "Antioxidante, Mejora la salud ocular", "10-30 mg/día"),
    ("Cianidina", "Cerezas, Frutas rojas", "Antioxidante, Mejora la salud vascular", "10-50 mg/día"),
    ("Delfinidina", "Arándanos, Grosellas", "Antioxidante, Mejora la memoria", "10-50 mg/día"),
    ("Petunidina", "Uvas negras, Moras", "Antioxidante, Mejora la salud cerebral", "10-30 mg/día"),
    ("Coumestrol", "Lentejas, Brotes", "Fitoestrógeno, Protector óseo", "10-30 mg/día"),
    ("Artemitina", "Hierbas amargas", "Antioxidante, Antimicrobiano", "10-30 mg/día"),
    ("Diosmetina", "Cítricos", "Antioxidante, Mejora la salud vascular", "10-30 mg/día"),
    ("Eriocitrina", "Limones", "Antioxidante, Mejora el metabolismo", "20-50 mg/día"),
    ("Limonina", "Cítricos", "Antioxidante, Protector hepático", "20-50 mg/día"),
    ("Nomilina", "Cáscaras de cítricos", "Antioxidante, Antimicrobiano", "10-30 mg/día"),
    ("Catequina gallato", "Té verde", "Antioxidante, Mejora la salud celular", "50-150 mg/día"),
    ("Flavan-3-oles", "Cacao, Uvas", "Antioxidante, Mejora la salud cardiovascular", "100-200 mg/día"),
    ("Prodelphinidinas", "Cacao, Moras", "Antioxidante, Mejora la circulación", "50-150 mg/día"),
    ("Ácido sinápico", "Mostaza, Granos", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Ácido p-hidroxibenzoico", "Cerezas, Manzanas", "Antioxidante, Antimicrobiano", "10-30 mg/día"),
    ("Ácido felúrico", "Cereales, Arroz", "Antioxidante, Protector dérmico", "20-50 mg/día"),
    ("Ácido tánico", "Té, Uvas", "Antioxidante, Mejora la digestión", "50-150 mg/día"),
    ("Glabridina", "Regaliz", "Antioxidante, Mejora la piel", "10-20 mg/día"),
    ("Ácido esculético", "Castaño de Indias", "Antioxidante, Mejora la circulación", "10-30 mg/día"),
    ("Aurantiamida", "Cáscaras de naranja", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Isoquercitrina", "Manzanas, Té verde", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Quercitrina", "Cebolla, Alcaparras", "Antioxidante, Mejora la salud vascular", "20-50 mg/día"),
    ("Sulfóxido de S-alilcisteína", "Ajo", "Antioxidante, Mejora la salud inmunológica", "5-20 mg/día"),
    ("Morina", "Moras, Cítricos", "Antioxidante, Mejora la memoria", "10-30 mg/día"),
    ("Artepillin C", "Propóleo", "Antioxidante, Antimicrobiano", "10-20 mg/día"),
    ("Humulona", "Lúpulo", "Antioxidante, Antiinflamatorio", "10-30 mg/día"),
    ("Xantohumol", "Lúpulo", "Antioxidante, Protector hepático", "20-50 mg/día"),
    ("Tetragidroxistilbeno", "Raíces de knotweed", "Antioxidante, Mejora la salud vascular", "10-30 mg/día"),
    ("Ácido anacárdico", "Anacardos", "Antioxidante, Antibacteriano", "5-15 mg/día"),
    ("Protocianidinas", "Cacao, Arándanos", "Antioxidante, Mejora la salud cardiovascular", "50-200 mg/día"),
    ("Calconoide", "Regaliz, Cítricos", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Eriodictiol", "Naranjas, Limones", "Antioxidante, Mejora el metabolismo", "20-50 mg/día"),
    ("Naringina", "Pomelo, Limones", "Antioxidante, Mejora la salud vascular", "50-150 mg/día"),
    ("Hesperetina", "Cítricos", "Antioxidante, Mejora la circulación", "20-50 mg/día"),
    ("Chalconas", "Regaliz, Lúpulo", "Antioxidante, Mejora la salud celular", "10-30 mg/día"),
    ("Flavonoles", "Cebolla, Té verde", "Antioxidante, Mejora la salud cerebral", "50-150 mg/día"),
    ("Robinetina", "Acacia", "Antioxidante, Mejora la memoria", "10-30 mg/día"),
    ("Grosularina", "Ribes", "Antioxidante, Mejora la salud ocular", "10-30 mg/día"),
    ("Pinobanksina", "Propóleo", "Antioxidante, Mejora la salud inmunológica", "10-30 mg/día"),
    ("Catequina epigalato", "Té verde", "Antioxidante, Protector neurológico", "50-200 mg/día"),
    ("Benzotropileno", "Hierbas medicinales", "Antioxidante, Mejora el sistema nervioso", "10-20 mg/día"),
    ("Epoxiresveratrol", "Uvas, Moras", "Antioxidante, Protector cardiovascular", "20-50 mg/día"),
    ("Isorhamnetina", "Perejil, Té verde", "Antioxidante, Mejora la salud vascular", "20-50 mg/día"),
    ("Elenolicósidos", "Olivos", "Antioxidante, Protector hepático", "10-30 mg/día"),
    ("Naringenina-7-rutinosido", "Naranjas", "Antioxidante, Mejora la salud ocular", "10-30 mg/día"),
    ("Epicatequina", "Té verde, Cacao", "Antioxidante, Mejora la salud vascular", "50-200 mg/día"),
    ("Episiringina", "Cereales, Granos", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Ácido protocatéquico", "Frutas, Legumbres", "Antioxidante, Antiinflamatorio", "20-50 mg/día"),
    ("Ácido clorogénico", "Café, Manzanas", "Antioxidante, Mejora el metabolismo", "50-150 mg/día"),
    ("Epigalloil catequina", "Té verde", "Antioxidante, Mejora la salud cerebral", "50-200 mg/día"),
    ("Acacetina", "Hierbas amargas", "Antioxidante, Mejora la salud vascular", "10-30 mg/día"),
    ("Baicaleína", "Scutellaria, Menta", "Antioxidante, Protector hepático", "10-30 mg/día"),
    ("Bergenina", "Hierbas medicinales", "Antioxidante, Mejora la salud inmunológica", "10-30 mg/día"),
    ("Daidzeína", "Soja, Tofu", "Fitoestrógeno, Protector óseo", "20-50 mg/día"),
    ("Formononetina", "Soja, Alfalfa", "Fitoestrógeno, Mejora la salud cardiovascular", "10-30 mg/día"),
    ("Genisteína", "Soja, Legumbres", "Fitoestrógeno, Protector celular", "50-150 mg/día"),
    ("Biochanina A", "Alfalfa, Legumbres", "Antioxidante, Mejora la salud vascular", "20-50 mg/día"),
    ("Orobol", "Soja", "Fitoestrógeno, Protector hepático", "10-30 mg/día"),
    ("Equol", "Derivado de isoflavonas", "Fitoestrógeno, Mejora la salud ósea", "10-20 mg/día"),
    ("Puerarina", "Raíz de pueraria", "Antioxidante, Mejora la circulación", "20-50 mg/día"),
    ("Gliciteína", "Soja, Tofu", "Fitoestrógeno, Mejora la salud hormonal", "20-50 mg/día"),
    ("Coumarina", "Canela, Cereza", "Antioxidante, Mejora la digestión", "10-20 mg/día"),
    ("Alfahidroxitiramina", "Aceitunas", "Antioxidante, Protector cardiovascular", "10-30 mg/día"),
    ("Tomentina", "Corteza de árbol", "Antioxidante, Mejora la piel", "10-30 mg/día"),
    ("Rutinósido", "Cítricos, Alforfón", "Antioxidante, Mejora la circulación", "20-50 mg/día"),
    ("Hidroxiviniferina", "Uvas", "Antioxidante, Mejora la salud vascular", "10-30 mg/día"),
    ("Resveratrol trimetilado", "Uvas, Moras", "Antioxidante, Mejora la memoria", "10-30 mg/día"),
    ("Betulin", "Abedul", "Antioxidante, Mejora la salud dérmica", "10-30 mg/día"),
    ("Licocalcona", "Raíz de regaliz", "Antioxidante, Mejora la salud hormonal", "10-20 mg/día"),
    ("Polidatina", "Uvas, Knotweed", "Antioxidante, Mejora la salud cerebral", "20-50 mg/día"),
    ("Oligómeros proantocianidínicos", "Cacao, Uvas", "Antioxidante, Mejora la salud cardiovascular", "50-200 mg/día"),
    ("Ácido dehidroascórbico", "Frutas cítricas", "Antioxidante, Protector celular", "20-50 mg/día"),
    ("Escopoletina", "Corteza de árbol", "Antioxidante, Mejora la circulación", "10-30 mg/día"),
    ("Glucohesperidina", "Naranjas", "Antioxidante, Mejora la salud ocular", "10-30 mg/día"),
    ("Kempferol 3-glucósido", "Té verde, Frutas", "Antioxidante, Mejora el metabolismo", "20-50 mg/día"),
    ("Miricetina", "Té, Uvas", "Antioxidante, Mejora la salud cerebral", "20-50 mg/día"),
    ("Teaflavina", "Té negro", "Antioxidante, Mejora la salud vascular", "50-100 mg/día"),
    ("Teaflavina digalato", "Té negro", "Antioxidante, Mejora la salud cardiovascular", "50-100 mg/día"),
    ("Proantocianidinas oligoméricas", "Cacao, Arándanos", "Antioxidante, Mejora la memoria", "50-150 mg/día"),
    ("Ácido genístico", "Té, Legumbres", "Antioxidante, Mejora la digestión", "20-50 mg/día"),
    ("Dimetilresveratrol", "Uvas", "Antioxidante, Mejora la circulación", "10-30 mg/día"),
    ("Artepillin B", "Propóleo", "Antioxidante, Antibacteriano", "10-30 mg/día"),
    ("Xantonas", "Mangostán, Hierbas", "Antioxidante, Mejora la salud inmune", "10-30 mg/día"),
    ("Geraldina", "Plantas herbáceas", "Antioxidante, Mejora la salud ocular", "10-30 mg/día"),
    ("Vitexina", "Tilo, Frutos", "Antioxidante, Mejora el metabolismo", "10-30 mg/día"),



]

# Insertar los datos si la tabla está vacía
cursor.executemany("INSERT INTO compuestos (nombre, alimentos, funcion, ingesta_recomendada) VALUES (?, ?, ?, ?)", compuestos)
conn.commit()

# Función para mostrar todos los compuestos
def mostrar_compuestos():
    cursor.execute("SELECT * FROM compuestos")
    rows = cursor.fetchall()
    print("\nBase de datos de compuestos:\n")
    print("{:<3} {:<20} {:<40} {:<50} {:<20}".format("ID", "Nombre", "Alimentos", "Función", "Ingesta Recomendada"))
    print("="*130)
    for row in rows:
        print("{:<3} {:<20} {:<40} {:<50} {:<20}".format(row[0], row[1], row[2], row[3], row[4]))

# Función para buscar un compuesto por nombre
def buscar_compuesto():
    nombre = input("\nIntroduce el nombre del compuesto a buscar: ").capitalize()
    cursor.execute("SELECT * FROM compuestos WHERE nombre LIKE ?", (f"%{nombre}%",))
    rows = cursor.fetchall()
    if rows:
        print("\nResultados:\n")
        print("{:<3} {:<20} {:<40} {:<50} {:<20}".format("ID", "Nombre", "Alimentos", "Función", "Ingesta Recomendada"))
        print("="*130)
        for row in rows:
            print("{:<3} {:<20} {:<40} {:<50} {:<20}".format(row[0], row[1], row[2], row[3], row[4]))
    else:
        print("\nNo se encontró el compuesto.")

# Función para crear recetas basadas en funcionalidad
def crear_receta():
    print("\nSeleccione una funcionalidad para generar recetas:")
    funcionalidades = list(set([compuesto[2] for compuesto in compuestos]))
    for i, funcionalidad in enumerate(funcionalidades):
        print(f"{i+1}. {funcionalidad}")
    opcion = int(input("\nIntroduce el número de tu elección: ")) - 1
    funcionalidad_elegida = funcionalidades[opcion]

    print(f"\nAlimentos sugeridos para la funcionalidad '{funcionalidad_elegida}':\n")
    cursor.execute("SELECT alimentos FROM compuestos WHERE funcion LIKE ?", (f"%{funcionalidad_elegida}%",))
    alimentos = cursor.fetchall()
    receta = [item[0] for item in alimentos]
    print(f"Alimentos sugeridos: {', '.join(set(receta))}")

# Menú principal
def menu():
    while True:
        print("\nOpciones:")
        print("1. Mostrar todos los compuestos")
        print("2. Buscar un compuesto")
        print("3. Crear receta basada en funcionalidad")
        print("4. Salir")
        opcion = input("\nElige una opción: ")

        if opcion == '1':
            mostrar_compuestos()
        elif opcion == '2':
            buscar_compuesto()
        elif opcion == '3':
            crear_receta()
        elif opcion == '4':
            print("\n¡Hasta luego!")
            break
        else:
            print("\nOpción no válida, intenta de nuevo.")

# Llamar al menú
menu()

# Cerrar la conexión
conn.close()
```
