```python
import pandas as pd

# Polifenoles de los ingredientes

ingredientes_cereales = {
    "Avena": {"quercetina": 1.5, "catequina": 2.0},
    "Arroz": {"dopamina": 1.2, "quercetina": 0.5},
}

ingredientes_aceites = {
    "Aceite de oliva": {"resveratrol": 0.8, "quercetina": 1.1},
    "Aceite de coco": {"dopamina": 1.3, "catequina": 2.2},
}

ingredientes_verduras = {
    "Espinaca": {"quercetina": 2.5, "catequina": 3.0},
    "Brócoli": {"dopamina": 1.5, "resveratrol": 1.0},
}

ingredientes_frutas = {
    "Uva pasa ": {
        "Quercetina 3-O-rutinósido": 0.51,
        "Ácido protocatecuico": 0.32,
        "Ácido 2-S-Glutationil caftárico": 0.47,
        "Ácido cafeoil tartárico": 4.96,
        "Ácido tartárico p-coumarol": 1.22,
        "Ácido tartárico trans-cafeoilo": 4.96,
        "Ácido tártaro trans-p-coumaroyl": 1.22,
    },
    "Albaricoque, seco": {
        "Lariciresinol": 0.06,
        "Mataresinol": 0.0001,
        "Pinoresinol": 0.19,
        "Secoisolariciresinol": 0.38,
    },
    "Ciruela, seca": {  # Nuevo ingrediente añadido
        "Quercetina 3-O-rutinósido": 2.50,
        "Ácido 3-cafeoilquino": 118.59,
        "Ácido 3-p-Coumaroilquínico": 1.32,
        "Ácido 4-cafeoilquínico": 31.25,
        "Ácido 5-cafeoilquino": 38.79,
        "Ácido cafeico": 1.11,
        "Ácido p-cumárico": 1.11,
    },
        "Dátiles, secos": {  # Nuevo ingrediente añadido
        "Ácido gálico": 1.56,
        "Ácido protocatecuico": 4.94,
        "Ácido sirrígico": 6.06,
        "Ácido de vainilla": 4.13,
        "Ácido cafeico": 2.52,
        "Ácido ferúlico": 11.83,
        "Ácido o-cumárico": 2.88,
        "Ácido p-cumárico": 5.77,
    },
        "Higos, secos": {  # Nuevo ingrediente añadido
        "Mataresinol": 0.001,  # Convertido de 1.00e-03 mg/100 g FW
        "Secoisolariciresinol": 0.11,
    },
        "Arándano europeo": {  # Nuevo ingrediente añadido
        "Miricetina": 6.96,
        "Quercetina": 16.39,
        "Ácido 4-hidroxibenzoico": 0.78,
        "Ácido 5-cafeoilquino": 5.09,
        "Ácido cafeico": 1.26,
        "Ácido ferúlico": 5.87,
        "Ácido p-cumárico": 10.12,
        "(-)-Epicatequina": 4.20,
        "Quercetina": 0.51,
        "Ácido p-cumárico": 2.02,
        "Resveratrol": 1.92,

    },
        "Arándano americano": {  # Nuevo ingrediente añadido
        "Cianidina 3-O-arabinósido": 4.47,
        "Cianidina 3-O-galactósido": 8.89,
        "Cianidina 3-O-glucósido": 0.74,
        "Peonidina 3-O-arabinósido": 9.61,
        "Peonidina 3-O-galactósido": 22.02,
        "Peonidina 3-O-glucósido": 4.16,
        "3-O-glucósido de Kaempferol": 0.87,
        "Miricetina 3-O-arabinósido": 5.30,
        "Quercetina 3-O-arabinosido": 4.94,
        "Quercetina 3-O-galactósido": 10.81,
        "Quercetina 3-O-ramnósido": 6.17,
        "Ácido 2,4-dihidroxibenzoico": 0.80,
        "Ácido 3-hidroxibenzoico": 0.41,
        "Ácido 4-hidroxibenzoico": 0.42,
        "Ácido benzoico": 48.10,
        "Ácido de vainilla": 2.81,
        "Ácido cafeico": 2.31,
        "Ácido cinámico": 0.16,
        "Ácido ferúlico": 0.81,
        "Ácido p-cumárico": 1.08,
       # Datos del arándano americano
        # Flavonoles
        "Kaempferol": 0.08,
        "Miricetina": 1.85,
        "Quercetina": 17.34,

        # Ácidos fenólicos - Ácidos hidroxibenzoicos
        "Ácido 2,3-dihidroxibenzoico": 0.31,
        "Ácido 2,4-dihidroxibenzoico": 4.25,
        "Ácido 2-hidroxibenzoico": 2.32,
        "Ácido 3-hidroxibenzoico": 0.91,
        "Ácido 4-hidroxibenzoico": 2.16,
        "Ácido benzoico": 174.10,
        "Ácido de vainilla": 1.92,

        # Ácidos fenólicos - Ácidos hidroxicinámicos
        "Ácido cafeico": 15.64,
        "Ácido cinámico": 2.05,
        "Ácido ferúlico": 8.79,
        "Ácido o-cumárico": 8.90,
        "Ácido p-cumárico": 25.38,
        "Ácido sinápico": 21.18,

        # Ácidos fenólicos - Ácidos hidroxifenilacéticos
        "Ácido 4-hidroxifenilacético": 0.73,
    },
        "Mora cruda": {  # Datos de la mora cruda
        # Flavonoides - Antocianinas
        "Cianidina 3-O-(6''-dioxalil-glucósido)": 10.97,
        "Cianidina 3-O-(6''-malonil-glucósido)": 4.30,
        "Cianidina 3-O-glucósido": 138.72,
        "Cianidina 3-O-rutinósido": 8.86,
        "Cianidina 3-O-xilósido": 9.74,
        # Antocianinas
        "Antocianinas": 246.8,
        # Flavonoides - Flavanoles
        "(+)-Catequina": 0.72,
        "(-)-Epicatequina": 11.48,
        "(-)-Epigalocatequina": 0.15,
        "Dímero de procianidina B1": 0.23,
        "Dímero de procianidina B2": 0.83,
        "Dímero de procianidina B3": 0.33,
        "Trímero de procianidina EEC": 0.06,
        "Dímero de prodelfinidina B3": 0.07,

        # Flavonoides - Flavonoles
        "Quercetina 3-O-galactósido": 4.10,
        "Quercetina 3-O-glucósido": 0.67,
        "Quercetina 3-O-glucosil-xilósido": 0.46,
        "Quercetina 3-O-rutinósido": 3.89,
        "Quercetina 3-O-xilósido": 1.28,
        "Quercetina 3-O-xilosil-glucurónido": 2.37,

        # Ácidos fenólicos - Ácidos hidroxibenzoicos
        "Ácido 4-hidroxibenzoico 4-O-glucósido": 1.13,
        "Ácido elágico": 43.67,
        "Ácido gálico": 4.67,
        "Galloyl glucose": 0.27,
        "Ácido protocatecúico 4-O-glucósido": 0.43,

        # Ácidos fenólicos - Ácidos hidroxicinámicos
        "Ácido 3-cafeoilquino": 4.53,
        "Ácido 3-feruloilquínico": 0.30,
        "Ácido 3-p-Coumaroilquínico": 0.37,
        "Ácido 4-cafeoilquínico": 0.10,
        "Ácido 5-cafeoilquino": 0.10,
        "Glucosa cafeoil": 0.50,
        "Feruloil glucosa": 0.43,
        "Ácido p-cumárico 4-O-glucósido": 0.27,
        "Glucosa p-coumaroil": 0.67,

        # Datos de la mora cruda
        # Flavonoides - Flavonoles
        "Kaempferol": 0.09,
        "Miricetina": 5.85,
        "Quercetina": 0.87,

        # Ácidos fenólicos - Ácidos hidroxibenzoicos
        "Ácido elágico": 31.91,
        "Ácido gálico": 5.27,

        # Ácidos fenólicos - Ácidos hidroxicinámicos
        "Ácido cafeico": 2.51,
        "Ácido ferúlico": 3.25,
        "Ácido p-cumárico": 1.24,

        # Lignanos
        "Ciclolariciresinol": 7.96e-3,
        "Lariciresinol": 0.15,
        "Lariciresinol-sesquilignan": 0.15,
        "Mataresinol": 9.07e-4,
        "Medioresinol": 0.05,
        "Pinoresinol": 0.14,
        "Secoisolariciresinol": 0.10,
        "Secoisolariciresinol-sesquilignan": 0.13,
        "Jeringaresinol": 0.19,
    },
    "Frambuesa roja": {
    # Flavonoides - Antocianinas
    "Cianidina": 0.53,
    "Cianidina 3-O-glucósido": 14.89,
    "Cianidina 3-O-glucosil-rutinósido": 7.06,
    "Cianidina 3-O-rutinósido": 5.20,
    "Cianidina 3-O-soforósido": 37.61,
    "Delphinidina 3-O-glucósido": 0.21,
    "Malvidin 3-O-glucósido": 0.62,
    "Pelargonidina 3-O-glucósido": 1.65,
    "Pelargonidina 3-O-glucosil-rutinósido": 0.82,
    "Pelargonidina 3-O-rutinósido": 0.42,
    "Pelargonidina 3-O-soforósido": 3.46,

    # Flavonoides - Flavanoles
    "(+)-Catequina": 0.58,
    "(-)-Epicatequina": 5.05,
    "Dímero de procianidina B2": 0.10,

    # Flavonoides - Flavonoles
    "Kaempferol": 0.00214,
    "3-O-glucósido de Kaempferol": 1.03,
    "Quercetina": 0.02,
    "Quercetina 3-O-glucósido": 3.58,
    "Quercetina 3-O-glucurónido": 0.63,
    "Quercetina 3-O-rutinósido": 11.00,

    # Ácidos fenólicos - Ácidos hidroxibenzoicos
    "Ácido elágico": 2.12,
    "Ácido elágico acetil-arabinósido": 0.20,
    "Ácido elágico acetil-xilósido": 0.36,
    "Ácido elágico arabinósido": 2.27,
    "Lambertianina C": 30.84,
    "Sanguiin H-6": 76.10,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 5-cafeoilquino": 0.57,
    "Ácido p-cumárico": 0.00023,
    "Ácido p-cumárico 4-O-glucósido": 0.32,
},
    "Grosella roja": {
    # Flavonoides - Antocianinas
    "Cianidina 3-O-glucósido": 3.37,
    "Cianidina 3-O-glucosil-rutinósido": 4.23,
    "Cianidina 3-O-rutinósido": 2.10,
    "Cianidina 3-O-sambubiosido": 9.47,
    "Cianidina 3-O-soforósido": 2.62,
    "Cianidina 3-O-xilosil-rutinósido": 11.22,
    "Delphinidina 3-O-sambubioside": 0.12,

    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.27,
    "(+)-Gallocatechina": 1.28,
    "(-)-Epicatequina": 0.08,
    "(-)-Epigalocatequina": 0.15,
    "Dimérico de procianidina B3": 0.20,
    "Dimero de prodelfinidina B3": 1.70,

    # Flavonoides - Flavonoles
    "Miricetina 3-O-glucósido": 0.06,
    "Quercetina 3-O-glucósido": 0.27,
    "Quercetina 3-O-rutinósido": 0.44,

    # Ácidos fenólicos - Ácidos hidroxibenzoicos
    "Ácido 4-hidroxibenzoico 4-O-glucósido": 1.03,
    "Ácido protocatecúico 4-O-glucósido": 0.07,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 3-p-Coumaroilquínico": 0.03,
    "Ácido 5-cafeoilquino": 0.10,
    "Ácido 5-feruloilquiníco": 0.17,
    "Ácido cafeico 4-O-glucósido": 0.13,
    "Glucosa cafeoil": 0.33,
    "Ácido cinámico": 0.10,
    "Ácido p-cumárico": 0.18,
    "Ácido p-cumárico 4-O-glucósido": 0.79,
    "glucosa p-coumaroil": 0.07,

    # Stilbenes
    "Resveratrol": 1.57,
    #Polifelnoles
    "polifenoles": 356,
},
    "Fresa, cruda": {
    # Flavonoides - Antocianinas
    "Cianidina": 0.50,
    "Cianidina 3-O-(6''-succinil-glucósido)": 1.22,
    "Cianidina 3-O-glucósido": 2.88,
    "Pelargonidina": 4.31,
    "Pelargonidina 3-O-(6''-malonil-glucósido)": 4.78,
    "Pelargonidina 3-O-(6''-succinil-glucósido)": 10.44,
    "Pelargonidina 3-O-arabinósido": 0.42,
    "Pelargonidina 3-O-glucósido": 47.14,
    "Pelargonidina 3-O-rutinósido": 1.32,

    # Flavonoides - Flavanoles
    "(+)-Catecina": 6.36,
    "(+)-Gallocatechina": 0.05,
    "(-)-Epicatequina": 0.0075,
    "(-)-Epicatequina 3-O-gallato": 0.28,
    "(-)-Epigalocatequina": 0.06,
    "Dímero de procianidina B1": 0.62,
    "Dímero de procianidina B2": 0.03,
    "Dimérico de procianidina B3": 1.10,
    "Dímero de procianidina B4": 0.13,
    "Trímero de procianidina EEC": 0.50,

    # Flavonoides - Flavonoles
    "3-O-glucósido de Kaempferol": 0.32,
    "3-O-glucurónido de Kaempferol": 0.20,
    "Morina": 0.06,
    "Quercetina 3-O-glucurónido": 1.74,

    # Ácidos fenólicos - Ácidos hidroxibenzoicos
    "Ácido 4-hidroxibenzoico 4-O-glucósido": 1.53,
    "Ácido 5-O-galoilquínico": 0.05,
    "Ácido elágico": 1.24,
    "Glucósido de ácido elágico": 2.85,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 5-cafeoilquino": 1.93,
    "Glucosa cafeoil": 0.10,
    "Ácido cinámico": 0.22,
    "Feruloil glucosa": 0.10,
    "Ácido p-cumárico": 0.21,
    "Ácido p-cumárico 4-O-glucósido": 0.15,
    "Glucosa p-coumaroil": 4.36,

    # Stilbenes
    "Resveratrol": 0.35,
},
    "Uva Verde": {
    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.41,
    "(+)-Gallocatechina": 0.01,
    "(-)-Epicatequina": 0.49,
    "(-)-Epicatequina 3-O-gallato": 0.25,
    "(-)-Epigalocatequina": 0.02,
    "Dímero de procianidina B1": 0.64,
    "Dímero de procianidina B2": 0.06,
    "Dímero de procianidina B3": 0.20,
    "Dímero de procianidina B4": 0.38,
    "Trímero de procianidina C1": 0.07,
    "Trímero de procianidina EEC": 0.13,
    "Dímero de prodelfinidina B3": 0.12,

    # Flavonoides - Flavonoles
    "Isorhamnetina 3-O-glucósido": 0.17,
    "Kaempferol 3-O-galactósido": 0.13,
    "3-O-glucósido de Kaempferol": 0.57,
    "Quercetina 3-O-glucurónido": 1.50,
    "Quercetina 3-O-rutinósido": 0.12,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido cafeoil tartárico": 2.74,
    "Ácido cis-cafeoil tartárico": 0.12,
    "Ácido tartárico Cis-p-Coumaroyl": 0.35,
    "Ácido tartárico p-coumarol": 1.00,
    "Ácido tartárico trans-cafeoilo": 5.23,
    "Ácido tártaro trans-p-coumaroyl": 1.45,

    # Stilbenes
    "Resveratrol": 0.02,
    "Resveratrol 3-O-glucósido": 0.24,
    "Trans-Resveratrol": 0.03,
    "Polifenoles": 112,
},
    "Uva Negra": {
    # Flavonoides - Antocianinas
    "Cianidina 3-O-(6''-p-coumaroyl-glucósido)": 0.10,
    "Cianidina 3-O-glucósido": 1.08,
    "Delphinidin 3-O-(6''-acetil-glucósido)": 0.54,
    "Delphinidina 3-O-glucósido": 2.63,
    "Malvidin 3-O-(6''-acetil-glucósido)": 9.66,
    "Malvidin 3-O-(6''-p-coumaroyl-glucósido)": 9.91,
    "Malvidin 3-O-glucósido": 39.23,
    "Peonidina 3-O-(6''-p-coumaroyl-glucósido)": 0.34,
    "Peonidina 3-O-glucósido": 5.80,
    "Petunidin 3-O-(6''-p-coumaroyl-glucósido)": 0.05,
    "Petunidina 3-O-glucósido": 2.76,

    # Flavonoides - Flavanoles
    "(+)-Catecina": 5.46,
    "(-)-Epicatequina": 5.24,
    "(-)-Epicatequina 3-O-gallato": 1.68,
    "(-)-Epigalocatequina": 0.03,
    "Dímero de procianidina B1": 0.43,
    "Dímero de procianidina B2": 0.36,
    "Dímero de procianidina B3": 0.12,
    "Dímero de procianidina B4": 0.33,
    "Trímero de procianidina C1": 0.38,

    # Flavonoides - Flavonoles
    "Quercetina 3-O-galactósido": 0.93,
    "Quercetina 3-O-glucurónido": 2.15,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido cafeoil tartárico": 1.13,
    "Ácido tartárico p-coumarol": 0.56,

    # Stilbenes
    "Piceatannol": 0.0052,
    "Resveratrol": 0.15,
    "Resveratrol 3-O-glucósido": 0.03,
    "Trans-Resveratrol": 0.15,
    "Trans-Resveratrol 3-O-glucósido": 0.01,
    "Polifenoles": 156,
},
    "Naranja": {
    # Flavonoides - Flavanonas
    "Hesperetina": 33.60,
    "Naringen": 11.22,

    # Flavonoides - Flavonoles
    "Kaempferol": 0.10,

    # Lignans
    "Lariciresinol": 0.08,
    "Mataresinol": 0.000661,
    "Medioresinol": 0.0095,
    "Pinoresinol": 0.02,
    "Secoisolariciresinol": 0.02,
    "Jeringaresinol": 0.12,
    "Polifenoles": 278,
},
    "Mandarina": {
    # Lignans
    "Lariciresinol": 0.06,
    "Mataresinol": 0.000231,
    "Pinoresinol": 0.02,
    "Secoisolariciresinol": 0.00646,
    "Polifenoles": 192
},
    "Pomelo": {
    # Ácidos fenólicos
    "Ácido 4-hidroxibenzoico": 0.005,
    "Ácido gálico": 0.006,
    "Ácido protocatecuico": 0.01,
    "Ácido de vainilla": 0.002,
    "Ácido cafeico": 0.002,
    "Ácido ferúlico": 0.01,
    "Ácido p-cumárico": 0.004,

    # Flavonoides
    "Hesperetina": 1.5,
    "Naringen": 53.00,

    # Flavonoles
    "Kaempferol": 0.27,
    "Quercetina": 0.33,

    # Lignans
    "Lariciresinol": 0.07,
    "Mataresinol": 0.000573,
    "Pinoresinol": 0.03,
    "Secoisolariciresinol": 0.02,




    # Ácidos fenólicos
    "Ácido 2-hidroxibenzoico": 0.26,
    "Ácido 4-hidroxibenzoico": 0.21,
    "Ácido gálico": 0.01,
    "Ácido protocatecuico": 0.04,
    "Ácido de vainilla": 0.44,
    "Ácido cafeico": 0.23,
    "Ácido ferúlico": 5.33,
    "Ácido p-cumárico": 0.77,
    "Ácido sinápico": 0.59,
    "Polifenoles": 25.2


},
    "Limón": {
    # Flavonoides
    "Eriodictiol": 17.60,
    "Hesperetina": 17.10,
    "Naringen": 0.55,

    # Flavones
    "Luteolina": 1.27,

    # Flavonoles
    "Quercetina": 0.37,

    # Lignans
    "Secoisolariciresinol": 0.02,
    "Polifenoles": 59.80,
},
    "Lima": {
    # Flavonoides
    "Hesperetina": 43.00,
    "Naringen": 3.40,

    # Flavonoles
    "Quercetina": 0.40,
},
    "Nectarina": {
    # Flavonoides
    "Cianidina 3-O-glucósido": 0.56,
    "(+)-Catecina": 4.78,
    "(-)-Epicatequina": 2.98,
    "Dimer de procianidina B1": 13.07,
    "Quercetina 3-O-galactósido": 0.58,

    # Ácidos fenólicos
    "Ácido 3-cafeoilquino": 5.13,
    "Ácido 5-cafeoilquino": 8.19,
    "Polifenoles": 47.32,
},
    "Melocotón, pelado": {
    # Flavonoides
    "Cianidina 3-O-glucósido": 0.28,
    "(+)-Catecina": 5.47,
    "(-)-Epicatequina": 7.97,
    "(-)-Epicatequina 3-O-gallato": 0.01,
    "Dimer de procianidina B1": 25.77,
    "Dimero de procianidina B2": 2.32,
    "Dimérico de procianidina B3": 2.44,
    "Dimer de procianidina B4": 0.13,
    "Dimero de procianidina B5": 0.04,
    "Dimero de procianidina B7": 1.30,
    "Trimer de procianidina C1": 2.53,
    "Trimer de procianidina EEC": 0.34,
    "Quercetina 3-O-galactósido": 0.71,

    # Ácidos fenólicos
    "Ácido 3-cafeoilquino": 4.13,
    "Ácido 5-cafeoilquino": 5.25,
    "Ácido cafeico": 0.63,

    # Lignans
    "Lariciresinol": 0.02,
    "Medioresinol": 3.50e-03,
    "Pinoresinol": 0.09,
    "Secoisolariciresinol": 1.00e-02,
    "Jeringaresinol": 1.50e-03,
    "Polifenoles": 106.02,
},
    "Melocotón, entero": {
    # Flavonoides
    "(+)-Catecina": 2.33,

    # Ácidos fenólicos
    "Ácido 3-cafeoilquino": 8.75,
    "Ácido 3-feruloilquiníco": 0.20,
    "Ácido 3-p-Coumaroilquínico": 0.35,
    "Ácido 5-cafeoilquino": 15.55,

    # Lignans
    "Lariciresinol": 0.06,
    "Mataresinol": 1.71e-04,
    "Pinoresinol": 0.15,
    "Secoisolariciresinol": 0.03,

    # Polifenoles
    "Polifenoles": 279.08,
},
    "Ciruela, fresca": {
    # Antocianinas
    "Cianidina 3-O-glucósido": 8.63,
    "Cianidina 3-O-rutinósido": 33.85,
    "Peonidina 3-O-glucósido": 0.46,
    "Peonidina 3-O-rutinósido": 4.85,

    # Flavanoles
    "(+)-Catecina": 4.60,
    "(-)-Epicatequina": 2.22,
    "Dimer de procianidina B1": 8.84,
    "Dimero de procianidina B2": 5.20,
    "Dimérico de procianidina B3": 1.00,
    "Dimer de procianidina B4": 1.02,
    "Dimero de procianidina B5": 1.59,
    "Dimero de procianidina B7": 4.69,
    "Trimer de procianidina C1": 10.01,
    "Trimer de procianidina EEC": 7.73,

    # Flavonoles
    "Quercetina 3-O-galactósido": 0.27,
    "Quercetina 3-O-glucósido": 0.54,
    "Quercetina 3-O-rutinósido": 5.90,

    # Ácidos fenólicos
    "Ácido 3-cafeoilquino": 75.88,
    "Ácido 3-feruloilquiníco": 1.85,
    "Ácido 3-p-Coumaroilquínico": 1.49,
    "Ácido 4-cafeoilquínico": 1.40,
    "Ácido 5-cafeoilquino": 8.40,

    # Flavonoles (otros)
    "Kaempferol": 0.06,
    "Miricetina": 0.06,
    "Quercetina": 2.10,

    # Ácidos fenólicos (hidroxibenzoicos)
    "Ácido protocatecuico": 0.07,
    "Ácido cafeico": 6.53,
    "Ácido ferúlico": 0.92,
    "Ácido p-cumárico": 0.91,

    # Lignans
    "Lariciresinol": 0.31,
    "Mataresinol": 2.22e-04,
    "Medioresinol": 1.00e-03,
    "Pinoresinol": 0.07,
    "Secoisolariciresinol": 0.01,

    # Polifenoles
    "Polifenoles, total": 409.79,
},
    "Cereza dulce, cruda": {
    # Flavonoides - Antocianinas
    "Cianidina 3-O-glucósido": 18.73,
    "Cianidina 3-O-rutinósido": 143.27,
    "Pelargonidina 3-O-rutinósido": 1.24,
    "Peonidina 3-O-glucósido": 0.76,
    "Peonidina 3-O-rutinósido": 7.42,

    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.50,
    "(-)-Epicatequina": 7.78,
    "(-)-Epicatequina 3-O-gallato": 0.09,
    "(-)-Epigalocatequina": 0.05,
    "Dímero de procianidina B1": 0.23,
    "Dímero de procianidina B2": 2.10,
    "Dímero de procianidina B3": 0.08,
    "Dímero de procianidina B4": 0.18,
    "Dímero de procianidina B5": 0.20,
    "Dímero de procianidina B7": 1.01,
    "Trímero de procianidina C1": 1.85,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 3-cafeoilquino": 44.71,
    "Ácido 3-feruloilquínico": 0.43,
    "Ácido 3-p-Coumaroilquínico": 38.43,
    "Ácido 4-cafeoilquínico": 0.77,
    "Ácido 4-p-Coumaroilquínico": 1.27,
    "Ácido 5-cafeoilquino": 2.20,

    # Flavonoles
    "Quercetina": 1.12,

    # Polifenoles
    "Polifenoles, total": 174.90,
},
    "Melón": {
    # Lignans
    "Lariciresinol": 1.80e-03,
    "Pinoresinol": 3.00e-04,
    "Secoisolariciresinol": 4.70e-03,

    # Polifenoles
    "Polifenoles, total": 109.59,
},
    "Sandia":{
    # Flavones
    "Luteolina": 1.84,

    # Lignans
    "Lariciresinol": 0.04,
    "Medioresinol": 1.00e-03,
    "Pinoresinol": 5.10e-03,
    "Secoisolariciresinol": 0.02,
    "Jeringaresinol": 0.02,

    # Polifenoles
    "Polifenoles, total": 51.44,
},
    "Manzana Sidra": {
    # Flavonoides - Dihidrocalcones
    "Floretina 2'-O-xilosil-glucósido": 3.28,
    "Phloridzin": 2.50,

    # Flavonoides - Flavanoles
    "(+)-Catecina": 5.56,
    "(-)-Epicatequina": 28.67,
    "Dimero de procianidina B1": 6.42,
    "Dimero de procianidina B2": 19.60,

    # Flavonoides - Flavonoles
    "Quercetina 3-O-galactósido": 8.29e-03,
    "Quercetina 3-O-glucósido": 0.06,
    "Quercetina 3-O-ramnósido": 0.22,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 4-p-Coumaroilquínico": 3.67,
    "Ácido 5-cafeoilquino": 49.33,
    "Ácido p-coumarolquino": 2.99,
    "Polifenoles": 250.89,

},
    "Manzana roja": {
    # Flavonoides - Dihidrocalcones
    "Floretina 2'-O-xilosil-glucósido": 0.46,
    "Phloridzin": 1.54,

    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.63,
    "(-)-Epicatequina": 6.72,
    "(-)-Epicatequina 3-O-gallato": 0.01,
    "Dimer de procianidina B1": 5.68,
    "Dimero de procianidina B2": 10.03,
    "Dimérico de procianidina B3": 0.23,
    "Dimero de procianidina B5": 0.97,
    "Dimero de procianidina B7": 3.76,
    "Trimer de procianidina C1": 7.02,
    "Trimer de procianidina EEC": 1.37,

    # Flavonoides - Flavonoles
    "Quercetina 3-O-arabinosido": 0.08,
    "Quercetina 3-O-galactósido": 0.16,
    "Quercetina 3-O-glucósido": 0.04,
    "Quercetina 3-O-ramnósido": 0.17,
    "Quercetina 3-O-rutinósido": 5.26e-05,
    "Quercetina 3-O-xilósido": 0.05,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 5-cafeoilquino": 18.22,
    "Feruloil glucosa": 0.15,
    "Glucosa p-coumaroil": 0.16,
    "Ácido p-coumarolquino": 1.38,

    # Lignans
    "Lariciresinol": 0.05,
    "Mataresinol": 4.42e-04,
    "Secoisolariciresinol": 1.38e-03,

    # Polifenoles totales
    "Polifenoles, total": 131.80,
},
    "Pera, pelada": {
    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.01,
    "(-)-Epicatequina": 1.43,
    "Dimer de procianidina B1": 1.00e-02,
    "Dimero de procianidina B2": 0.66,
    "Dimérico de procianidina B3": 5.00e-03,
    "Dimero de procianidina B5": 5.00e-03,
    "Dimero de procianidina B7": 0.05,
    "Trimer de procianidina C1": 0.24,
    "Trimer de procianidina EEC": 0.01,

    # Ácidos fenólicos - Ácidos hidroxicinámicos
    "Ácido 5-cafeoilquino": 3.18,
    "Ácido cafeico": 0.14,
    "Ácido p-cumárico": 0.18,

    # Otros polifenoles
    "Arbutin": 1.40,

    # Lignans
    "Lariciresinol": 0.02,
    "Mataresinol": 2.67e-05,
    "Medioresinol": 3.00e-03,
    "Pinoresinol": 1.00e-03,
    "Secoisolariciresinol": 3.11e-03,
    "Jeringaresinol": 5.50e-03,

    # Polifenoles totales
    "Polifenoles, total": 126.42,
},
    "Plátano": {
    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.34,
    "(-)-Epicatequina": 0.11,
    "(-)-Epigalocatequina": 2.14e-03,
    "Dimero de prodelfinidina B3": 0.10,

    # Ácidos fenólicos - Ácidos hidroxibenzoicos
    "Ácido gálico": 1.00,

    # Lignans
    "Lariciresinol": 2.20e-03,
    "Mataresinol": 5.45e-05,
    "Pinoresinol": 4.00e-05,
    "Secoisolariciresinol": 7.73e-05,
    "Jeringaresinol": 1.00e-02,

    # Polifenoles totales
    "Polifenoles, total": 154.70,
},
    "Kiwi": {
    # Flavonoides - Flavanoles
    "(+)-Catecina 3-O-galato": 0.08,
    "(-)-Epicatequina": 0.29,
    "(-)-Epigalocatequina 3-O-galato": 0.08,
    "Dimero de procianidina B2": 0.14,
    "Trimer de procianidina C1": 0.11,

    # Flavonoides - Flavones
    "Luteolina": 0.74,

    # Lignans
    "Lariciresinol": 0.02,
    "Mataresinol": 1.93e-03,
    "Medioresinol": 4.50e-03,
    "Pinoresinol": 3.60e-03,
    "Secoisolariciresinol": 0.11,
    "Jeringaresinol": 4.00e-03,

    # Polifenoles totales
    "Polifenoles, total": 179.71,
},
    "Mango": {
    # Flavonoides - Flavanoles
    "(+)-Catecina": 1.72,

    # Lignans
    "Mataresinol": 1.06e-03,
    "Secoisolariciresinol": 1.00e-02,

    # Polifenoles totales
    "Polifenoles, total": 144.77,
},
    "Papaya": {
    # Lignans
    "Mataresinol": 2.00e-03,

    # Polifenoles totales
    "Polifenoles, total": 57.60,
},
    "Maracuyá": {
    # Lignans
    "Secoisolariciresinol": 0.02,  # mg/100 g FW

    # Polifenoles totales
    "Polifenoles, total": 57.40,  # mg/100 g FW
},
    "Piña": {
    # Lignans
    "Lariciresinol": 0.02,  # mg/100 g FW
    "Mataresinol": 8.50e-03,  # mg/100 g FW
    "Medioresinol": 2.00e-03,  # mg/100 g FW
    "Pinoresinol": 4.40e-03,  # mg/100 g FW
    "Secoisolariciresinol": 4.50e-03,  # mg/100 g FW
    "Jeringaresinol": 0.09,  # mg/100 g FW

    # Polifenoles totales
    "Polifenoles, total": 147.91,  # mg/100 g FW
},
    "Granada": {
    # Flavonoides

        "(+)-Catecina": 0.40,  # mg/100 g FW
        "(+)-Gallocatechina": 0.17,  # mg/100 g FW
        "(-)-Epicatequina": 0.08,  # mg/100 g FW
        "(-)-Epigalocatequina": 0.16,  # mg/100 g FW
        "Dimer de procianidina B1": 0.13,  # mg/100 g FW
        "Dimérico de procianidina B3": 0.16,  # mg/100 g FW


    # Lignans

        "Mataresinol": 9.00e-03,  # mg/100 g FW
        "Secoisolariciresinol": 0.29,  # mg/100 g FW

}































    }


ingredientes_condimentos = {
    "Pimienta": {"quercetina": 1.0, "dopamina": 0.7},
    "Cúrcuma": {"resveratrol": 0.5, "catequina": 2.3},
}

ingredientes_semillas = {
    "Chía": {"quercetina": 2.0, "catequina": 1.8},
    "Linaza": {"dopamina": 1.0, "resveratrol": 1.5},
}

ingredientes_bebidas = {
    "Té verde": {"catequina": 3.2, "dopamina": 1.1},
    "Café": {"quercetina": 1.0, "resveratrol": 1.5},
}






















categorias = {
    1: ("Cereales", ingredientes_cereales),
    2: ("Aceites", ingredientes_aceites),
    3: ("Verduras", ingredientes_verduras),
    4: ("Frutas", ingredientes_frutas),  # Asegúrate de definir la lista de ingredientes para frutas
    5: ("Condimentos", ingredientes_condimentos),
    6: ("Semillas", ingredientes_semillas),
    7: ("Bebidas", ingredientes_bebidas),
}









# Función para seleccionar ingredientes y calcular polifenoles consumidos
def seleccionar_ingredientes(ingredientes_categoria):
    """Permite seleccionar ingredientes y calcular polifenoles consumidos."""
    polifenoles_consumidos = {}
    ingredientes_consumidos = {}

    while True:
        print("\nIngredientes disponibles:")
        for i, ingrediente in enumerate(ingredientes_categoria, 1):
            print(f"{i}. {ingrediente}")

        seleccion = input("Selecciona un ingrediente por su número (o 'fin' para terminar de seleccionar ingredientes): ")

        if seleccion.lower() == "fin":
            break

        try:
            seleccion = int(seleccion) - 1
            if seleccion < 0 or seleccion >= len(ingredientes_categoria):
                print("Selección inválida. Intenta de nuevo.")
                continue

            ingrediente = list(ingredientes_categoria.keys())[seleccion]
            cantidad = float(input(f"Ingresa la cantidad de {ingrediente} en gramos: "))

            # Acumular los polifenoles consumidos
            if ingrediente not in ingredientes_consumidos:
                ingredientes_consumidos[ingrediente] = 0
            ingredientes_consumidos[ingrediente] += cantidad

            for polifenol, cantidad_mg in ingredientes_categoria[ingrediente].items():
                if polifenol not in polifenoles_consumidos:
                    polifenoles_consumidos[polifenol] = 0
                polifenoles_consumidos[polifenol] += (cantidad * cantidad_mg) / 100

            print(f"Has seleccionado {cantidad} gramos de {ingrediente}.")
        except ValueError:
            print("Por favor, ingresa un número válido.")

        # Preguntar si desea continuar o terminar
        continuar = input("\n¿Quieres añadir otro ingrediente? (c para continuar, s para salir, fin para terminar): ")
        if continuar.lower() == 's':
            break
        elif continuar.lower() == 'fin':
            break

    return polifenoles_consumidos, ingredientes_consumidos

# Función para mostrar opciones después de seleccionar ingredientes
def mostrar_opciones(polifenoles_consumidos, ingredientes_consumidos):
    """Muestra opciones después de seleccionar ingredientes."""
    while True:


        print("1. Suma total de antioxidantes consumidos")

        print("2. Volver al inicio")
        opcion = input("Selecciona una opción (1/2): ")



        if opcion == "1":
            print("\nSuma total de polifenoles consumidos en mg:")
            total_polifenoles = 0  # Nueva variable para la suma total
            for polifenol, cantidad_total in polifenoles_consumidos.items():
                print(f" - {polifenol}: {cantidad_total:.2f} mg")
                total_polifenoles += cantidad_total  # Acumulador de la suma
            print(f"\nPolifenoles totales consumidos: {total_polifenoles:.2f} mg")
            # Mostrar como DataFrame
            data = {
                "Polifenol": [],
                "Cantidad consumida (mg)": []
            }
            for polifenol, cantidad in polifenoles_consumidos.items():
                if cantidad > 0:
                    data["Polifenol"].append(polifenol)
                    data["Cantidad consumida (mg)"].append(round(cantidad, 2))
            df = pd.DataFrame(data)
            print(df.to_string(index=False))



        elif opcion == "2":
            print("\nRegresando al inicio del programa...\n")
            return  # Regresa al inicio del programa

        else:
            print("Opción inválida. Por favor, elige 1 o 2.")

# Función para mostrar el menú de inicio
def mostrar_menu_inicio():
    """Muestra el menú de inicio con las categorías."""
    while True:
        print("\nSelecciona una categoría de ingredientes:")
        for key, (categoria, _) in categorias.items():
            print(f"{key}. {categoria}")
        print("8. Salir")

        seleccion = input("Selecciona una opción (1-8): ")

        if seleccion.isdigit():
            seleccion = int(seleccion)

            if seleccion == 8:
                print("Saliendo del programa...")
                break  # Salir del programa
            elif seleccion in categorias:
                categoria_nombre, ingredientes_categoria = categorias[seleccion]
                print(f"\nHas seleccionado la categoría: {categoria_nombre}")
                polifenoles_consumidos, ingredientes_consumidos = seleccionar_ingredientes(ingredientes_categoria)
                mostrar_opciones(polifenoles_consumidos, ingredientes_consumidos)
            else:
                print("Selección inválida. Intenta de nuevo.")
        else:
            print("Por favor, ingresa un número válido.")

# Programa principal
def jugar():
    mostrar_menu_inicio()

# Ejecutar el juego
jugar()
```
